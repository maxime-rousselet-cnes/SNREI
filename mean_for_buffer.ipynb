```python
from pandas import read_csv
from pathlib import Path

df = read_csv(Path(".").joinpath("data/results/tables/harmonic_load_signal_trends.csv"))
```

```python
from snrei.utils import harmonic_load_signal_trends_path, load_complex_array_from_binary

for id in [24, 74, 124, 174, 224, 274, 324, 374, 424]:
    a = load_complex_array_from_binary(name=str(id), path=harmonic_load_signal_trends_path.joinpath("step_1"))
    b = load_complex_array_from_binary(name=str(id), path=harmonic_load_signal_trends_path.joinpath("step_2"))
    c=a - b
    c[:, 1, :]*=0
    print(sum(abs(c).flatten()))
```

```python
from matplotlib.pyplot import plot, grid, show,  figure

for Love_id in [0, 1, 2]:
    figure(figsize=(10, 10))
    lines = df[df["Love_numbers_ID"] == Love_id][df["leakage_correction_iterations"] == 1][
        df["signal_threshold"] == 8.0
    ][df["signal_threshold_past"] == 8.0].sort_values(by="buffer_distance")
    print(lines["ID"])
    for mean in ["initial", "step_0", "step_1", "step_2", "step_3"]:
        plot(lines["buffer_distance"].values, lines["ocean_mean_" + mean].values)
    grid()
    show()
```
