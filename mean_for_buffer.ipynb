```python
from pandas import read_csv
from pathlib import Path

df = read_csv(Path(".").joinpath("data/results/tables/harmonic_load_signal_trends.csv"))
```

```python
from matplotlib.pyplot import plot, grid, show, figure, legend

for Love_id in [0, 1]:
    figure()
    lines = df[df["Love_numbers_ID"] == Love_id].sort_values(by="buffer_distance")
    for mean in ["step_1", "step_2", "step_3", "step_4", "step_5"]:
        plot(lines["buffer_distance"].values, lines["ocean_mean_" + mean].values, label=mean)
    legend()
    grid()
    show()
```
